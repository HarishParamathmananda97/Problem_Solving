```python
import re

text_to_search = """
abcdefghijklmnopqrstuvwxyz\n
ABCDEFGHIJKLMNOPQRSTUVWXYZ\n
1234567890

Ha HaHa

MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )
https://www.google.com
http://coreyms.com
https://youtube.com
https://www.nasa.gov
https://www.lifealwayshappy.com

coreyMSchafer@gmail.com
corey.schafer@university.edu
corey-321-schafer@my-work.net
jayanthikrish96@gmail.com

321-555-4321
123.555.1234
123*555*1234
800-555-1234
900-555-1234

Mrs. JayanthiHarish
Mr Harish
Mr. Babji
Mrs. Savitha Babji
Mrs. SriLakshmi Paramathmananda
Mr. Paramathmananda
Ms. Vishwa Babji

cat
bat
pat
mat
sat
tat


"""

sentence = 'Start a sentence and then bring it to an end.'

urls = """
https://www.google.com
http://coreyms.com
https://youtube.com
https://www.nasa.gov
"""
```

```python
""" 
Types of characters 
.     -  Any Character Except New Line
\d    -  Digit (0-9)
\D    -  Not a Digit (0-9)
\w    -  Word character (a-z, A-Z, 0-9, _)
\W    -  Not a Word Character.
\s    -  Whitespace (space, tab, newline)
\S    -  Not Whitespace (space, tab, newline)
\b    -  Word Boundary
\B    -  Not a Word Boundary
^     -  Begining of a String
$     -  End of a String

[]    -  Matches characters in brackets
[^ ]  -  Matches characters Not in 

brackets
|     -  Either OR
( )   -  Group

Quantifiers:
*     -  0 or More
+     -  1 or More
?     -  0 or One
{3}   -  Exact Number
{3, 4}-  Range of Numbers (Minimum, Maximum)

##### Sample Regex #####

[a-zA-Z0-9_.+-]+@[a-zA-Z0-9-]+\.[]
"""
```

```python
pattern = re.compile(r'abc')
pattern = re.compile(r'.')

# search through our text using this pattern.

matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
#r'\d' - matches only digits and not other character
pattern = re.compile(r'\d')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
#r'\D' - matches any character other than digits.
pattern = re.compile(r'\D')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
#r'\w' - word character (a-z, A-Z, 0-9, _) 
pattern = re.compile(r'\w')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
#r'\W' - word character (a-z, A-Z, 0-9, _) 
pattern = re.compile(r'\W')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
#r'\s' - selects spaces, tabs, and newlines(\n)

pattern = re.compile(r'\s')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
#r'\S' - Matches other than spaces, tabs, and newlines(\n) chars.

pattern = re.compile(r'\S')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
#r'\b' - Word Boundary  - words with starting and not middle word. HaHa
#_
pattern = re.compile(r'\bHa')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
#r'\b' - Not a Word Boundary  - words with starting and not middle word. 
#HaHa
#  _
pattern = re.compile(r'\BHa')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# ^ - Beginning of a String.
# $ - End of a String.
print(sentence)
print('----------------------------------------------')

pattern = re.compile(r'^Start')

matches = pattern.finditer(sentence)

for match in matches:
    print(match)  # gives one match

print('----------------------------------------------')
pattern = re.compile(r'end.$') # matches the end of the string

matches = pattern.finditer(sentence)

for match in matches:
    print(match)

print('----------------------------------------------')
```

```python
# challenge is we should select all the three characters like cat, mat, pat, etc... but shouldn't select bat.
# [^ ]  -  Matches characters Not in [^b ] will skip the letters starting with b.

pattern = re.compile(r'[^b]at')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
#pattern to match the prefix like Mr. , etc., and their name

import re

# pattern = re.compile(r'Mr\.?\s[A-Z]\w* ')
pattern = re.compile(r'(Mr|Ms|Mrs)\.?\s[A-Z]\w*')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```
