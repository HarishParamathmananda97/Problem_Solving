```python
import pandas as pd

# Read the CSV file and set 'rank' as index column
df = pd.read_csv('Fortune_1000.csv', index_col='rank')

# Print the first 5 rows of the DataFrame
df.head()
```

```python
sector_groups = df.groupby('sector')
sector_groups
```

```python
sector_groups.size()
```

```python
tech_group = sector_groups.get_group('Technology')
tech_group
```

```python
# Get the highest revenue company in each sector
highest_revenue = sector_groups.apply(lambda x: x.loc[x['revenue'].idxmax()])

# Print the result
highest_revenue
```

```python
highest_revenue_profit = sector_groups.apply(lambda x: x.loc[x['revenue'].idxmax()][['company', 'revenue', 'profit']])
highest_revenue_profit
```

```python
highest_profit = sector_groups.apply(lambda x: x.loc[x['profit'].idxmax()][['company', 'revenue', 'profit']])
highest_profit
```

```python
sector_state_groups = df.groupby(['sector', 'state'])
sector_state_groups
```

```python
# Get the highest profit company in each sector-state group
highest_profit = sector_state_groups.apply(lambda x: x.loc[x['profit'].idxmax()][['company', 'revenue', 'profit']])

# Print the result
highest_profit
```

```python
# Get the sum of revenue for each sector-state group
revenue_sum = sector_state_groups.agg({'revenue': 'sum'})

# Print the result
revenue_sum
```

```python
# Get the sum of revenue for each sector-state group
revenue_count = sector_state_groups.agg({'revenue': 'count'})

# Print the result
revenue_count
```
