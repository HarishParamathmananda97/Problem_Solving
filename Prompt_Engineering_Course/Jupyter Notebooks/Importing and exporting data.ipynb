```python
import pandas as pd

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

# Read the table from the Wikipedia page
df = pd.read_html(url, header=0)[0]

# Print the first 5 rows of the table
df.head()
```

```python
# Export the table to an Excel file
df.to_excel("S&P500.xlsx", index=False)
```

```python
# Export the table to a CSV file
df.to_csv("S&P500.csv", index=False)
```

```python
# Export the table to a TXT file
df.to_csv("S&P500.txt", index=False, sep="\t")
```

```python
import os
import pandas as pd

# Folder path where the CSV files are stored
folder_path = "/Users/saadahmed/Desktop/PandasDataAnalysis/Restaurant_data"

# List all the CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith(".csv")]

# Create a separate DataFrame for each CSV file
dfs = {}
for file in csv_files:
    file_name = os.path.splitext(file)[0].replace(" ", "_")
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    dfs[file_name] = df

# Print the first 5 rows of each DataFrame
for df_name, df in dfs.items():
    print("First 5 rows of", df_name)
    print(df.head())
    print()
```
