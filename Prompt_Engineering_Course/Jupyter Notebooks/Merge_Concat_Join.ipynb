```python
# Import necessary libraries
import pandas as pd

# Load data files into DataFrames
week1_sales = pd.read_csv('Restaurant - Week 1 Sales.csv')
week1_satisfaction = pd.read_csv('Restaurant - Week 1 Satisfaction.csv')
week2_sales = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')

```

```python
combined_sales = pd.concat([week1_sales, week2_sales])
combined_sales
```

```python
# Concatenate week1_sales and week2_sales DataFrames
combined_sales = pd.concat([week1_sales, week2_sales], ignore_index=True)
combined_sales
```

```python
# Combine week1_sales and week2_sales DataFrames using append
combined_sales = week1_sales.append(week2_sales, ignore_index=True)
combined_sales
```

```python
# Perform inner join on week1_sales and week2_sales DataFrames using 'Customer ID' as the key
merged_sales = week1_sales.merge(week2_sales, on='Customer ID', suffixes=('_week1', '_week2'))
merged_sales
```

```python
# Filter rows with 'Customer ID' equal to 155
filtered_customers = merged_sales[merged_sales['Customer ID'] == 155]
filtered_customers
```

```python
# Perform inner join on week1_sales and week2_sales DataFrames using 'Customer ID' and 'Food ID' as the keys
merged_sales = week1_sales.merge(week2_sales, on=['Customer ID', 'Food ID'], suffixes=('_week1', '_week2'))
merged_sales
```

```python
# Filter rows with 'Customer ID' equal to 21
filtered_customers = merged_sales[merged_sales['Customer ID'] == 21]
filtered_customers
```

```python
# Perform outer join on week1_sales and week2_sales DataFrames using 'Customer ID' and 'Food ID' as the keys
merged_sales_outer = week1_sales.merge(week2_sales, on=['Customer ID', 'Food ID'], how='outer', suffixes=('_week1', '_week2'), indicator=True)
merged_sales_outer
```

```python
# Filter rows with '_merge' equal to 'both'
both_weeks = merged_sales_outer[merged_sales_outer['_merge'] == 'both']
both_weeks
```

```python
# Perform left join on week1_sales and foods DataFrames using 'Food ID' as the key
week1_sales_with_food_info = week1_sales.merge(foods, left_on='Food ID', right_on='Food ID', how='left')
week1_sales_with_food_info
```

```python
# Perform right join on week1_sales_with_food_info and customers DataFrames using 'Customer ID' and 'ID' as the keys
week1_sales_complete_info = week1_sales_with_food_info.merge(customers, left_on='Customer ID', right_on='ID', how='left')
week1_sales_complete_info
```

```python
week1_sales = week1_sales.set_index(["Customer ID", "Food ID"])
week2_sales = week2_sales.set_index(["Customer ID", "Food ID"])
```

```python
combined_sales = week1_sales.join(week2_sales, lsuffix='_week1', rsuffix='_week2')
combined_sales
```
