```python
# Import necessary libraries
import yfinance as yf

# Define the ticker symbol
ticker_symbol = 'AAPL'

# Get data on the specified ticker
ticker_data = yf.Ticker(ticker_symbol)

# Get the historical prices for the specified period
hist_data = ticker_data.history(period='1d', start='2020-01-01', end='2023-01-01')

# Display the historical data
hist_data
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Extract the 'Close' column from the historical data DataFrame
closing_prices = hist_data['Close']

# Plot the closing prices
plt.plot(closing_prices)

# Set the title and labels for the plot
plt.title('Apple Price')
plt.xlabel('Month - Year')
plt.ylabel('Closing Price')


# Display the plot
plt.show()
```

```python
import numpy as np

# Create a new figure and set the size
plt.figure(figsize=(10, 6))

# Plot the closing prices with a blue solid line, markers, and a custom line width
plt.plot(closing_prices, color='blue', marker='o', linestyle='-', linewidth=1.5, label='AAPL Close Price')

# Set the title and labels for the plot
plt.title('Apple Inc. (AAPL) Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')

# Customize x-axis ticks to display one tick per year
years = np.arange(closing_prices.index.min().year, closing_prices.index.max().year + 1)
tick_positions = [f'{year}-01-01' for year in years]
tick_labels = [str(year) for year in years]
plt.xticks(tick_positions, tick_labels)

# Customize y-axis ticks with a step of 50
y_tick_positions = np.arange(0, closing_prices.max() + 50, 50)
plt.yticks(y_tick_positions)

# Add a legend and grid
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd

# Calculate the percentage change in closing prices
percentage_change = closing_prices.pct_change() * 100

# Create a new figure and set the size
plt.figure(figsize=(10, 6))

# Set the background color to purple
plt.gca().set_facecolor('purple')

# Plot the percentage change with a white solid line, smaller markers, and a custom line width
plt.plot(percentage_change, color='white', marker='o', markersize=3, linestyle='-', linewidth=2, label='AAPL % Change')

# Set the title and labels for the plot
plt.title('Apple Inc. (AAPL) Closing Price Percentage Change')
plt.xlabel('Date')
plt.ylabel('Percentage Change')

# Customize x-axis ticks and y-axis ticks as before
month_range = pd.date_range(closing_prices.index.min(), closing_prices.index.max(), freq='M')
tick_positions = month_range.strftime('%Y-%m-%d').tolist()
tick_labels = month_range.strftime('%Y-%m').tolist()
plt.xticks(tick_positions, tick_labels, rotation=90)
y_tick_positions = np.arange(percentage_change.min() - 5, percentage_change.max() + 5, 5)
plt.yticks(y_tick_positions)

# Add a legend (no grid)
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd

# Resample the closing prices by month and calculate the percentage change
monthly_closing_prices = closing_prices.resample('M').last()
monthly_percentage_change = monthly_closing_prices.pct_change() * 100

# Create a new figure and set the size
plt.figure(figsize=(10, 6))

# Set the background color to purple
plt.gca().set_facecolor('purple')

# Plot the monthly percentage change with a white solid line, smaller markers, and a custom line width
plt.plot(monthly_percentage_change, color='white', marker='o', markersize=3, linestyle='-', linewidth=2, label='AAPL % Change (Monthly)')

# Set the title and labels for the plot
plt.title('Apple Inc. (AAPL) Monthly Closing Price Percentage Change')
plt.xlabel('Date')
plt.ylabel('Percentage Change')

# Customize x-axis ticks and y-axis ticks as before
month_range = pd.date_range(closing_prices.index.min(), closing_prices.index.max(), freq='M')
tick_positions = month_range.strftime('%Y-%m-%d').tolist()
tick_labels = month_range.strftime('%Y-%m').tolist()
plt.xticks(tick_positions, tick_labels, rotation=90)
y_tick_positions = np.arange(monthly_percentage_change.min() - 5, monthly_percentage_change.max() + 5, 5)
plt.yticks(y_tick_positions)

# Add a legend (no grid)
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd

# Define the bins for percentage change groups
bins = [-np.inf, -10, 0, 10, np.inf]

# Group the percentage changes into the specified buckets
grouped_percentage_change = pd.cut(monthly_percentage_change, bins)

# Calculate the count of each bucket
bucket_counts = grouped_percentage_change.value_counts().sort_index()

# Create the histogram
plt.figure(figsize=(10, 6))
plt.gca().set_facecolor('purple')
bucket_counts.plot(kind='bar', color='white', edgecolor='white', width=0.8)

# Set the title and labels for the histogram
plt.title('Apple Inc. (AAPL) Monthly Closing Price Percentage Change Histogram')
plt.xlabel('Percentage Change Group')
plt.ylabel('Frequency')

# Customize x-axis ticks
plt.xticks(range(len(bucket_counts.index)), ['<-10%', '-10% to 0%', '0% to 10%', '>10%'], rotation=0)

# Customize y-axis ticks
y_tick_positions = np.arange(0, bucket_counts.max() + 1, 1)
plt.yticks(y_tick_positions)

# Display the histogram
plt.show()
```
